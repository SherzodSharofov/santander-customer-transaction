```python
import requests
import pandas as pd
import numpy as np

test_data = pd.read_csv('/content/test.csv')
test_data_ids = test_data['ID_code'].copy()
test_data = test_data.drop(columns='ID_code')
test_data = test_data.replace([np.inf, -np.inf], np.nan)
test_data = test_data.fillna(0)

url = "https://nonboiling-patt-nikolai.ngrok-free.dev/predict_batch"
all_predictions = []
chunk_size = 1000

for i in range(0, len(test_data), chunk_size):
    chunk = test_data.iloc[i:i+chunk_size]
    response = requests.post(
        url,
        json={"rows": chunk.to_dict(orient='records')},
        headers={"ngrok-skip-browser-warning": "true"}
    )
    result = response.json()
    all_predictions.extend(result['predictions'])
    print(f"processed {min(i+chunk_size, len(test_data))}/{len(test_data)}")

submission = pd.DataFrame({
    'ID_code': test_data_ids,
    'target': all_predictions
})
print(submission.head())
print(submission['target'].value_counts())
submission.to_csv('submission.csv', index=False)
```
